```python
import numpy as np
import time

from matrix_loader import *
from sklearn.svm import SVR
```

```python
svm_reg = SVR(verbose=True)
print("Loading data ...")
m = Matrix_loader()
X_train, y_train, X_val, y_val = m.load_raw_mats()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ex is the example number, in the 7k examples in the training set
# t is the time range (0:40) in 2 second intervals
#   time_step controls how many graphs to generate
# plot title is the Y target

time_step = 5
ex_step = 500

for ex in range(0, 7174, ex_step):   
    # time range
    gs = gridspec.GridSpec(2,(40/time_step))
    for t in range(0, 40, time_step):
        img = []
        # number of rows
        for r in range(13):
            img.append([X_train[ex][(40*13*r)+t+(40*c)] for c in range(0,13)])
        img = np.array(img)
        ax = plt.subplot(gs[0,t/time_step])
        plt.imshow(img, interpolation='nearest')

    aif = np.array(X_train[ex][-40:])
    elapsed_time = np.array([x for x in range(0, 80, 2)])
    ax = plt.subplot(gs[1,:])
    ax.set_title(y_train[ex])
    plt.plot(elapsed_time, aif, linewidth=0.2)
    plt.show()
```

```python
import matplotlib.pyplot as plt

ex = 0
# time range
aif = np.array(X_train[ex][-40:])
elapsed_time = np.array([x for x in range(0, 80, 2)])
plt.plot(elapsed_time, aif, linewidth=0.2)
plt.show()
```

```python
print("Training SVM ...")
start_time = time.time()
svm_reg.fit(X_train, y_train)
print("Training took {:.3f}s".format(time.time() - start_time))
```

```python
y_pred = svm_reg.predict(X_val)
print("Accuracy: {:.4f}".format(svm_reg.score(y_val, y_pred)))
```
