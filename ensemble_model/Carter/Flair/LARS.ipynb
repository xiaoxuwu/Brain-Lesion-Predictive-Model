```python
# import packages
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.externals import joblib
from Utils import matrix_loader as MatrixLoader
from sklearn.decomposition import PCA
import scipy.sparse as sp
from sklearn.neighbors import KNeighborsRegressor
```

```python
# manfiest constants
PATCH_RADIUS = 4

# load files into matrix
matrixLoader = MatrixLoader.Matrix_loader('/home/carter/Documents/CS-188-Ensemble-ML-Model/Data')
(trainA, trainb, verA, verb) = matrixLoader.load(PATCH_RADIUS)

# compute width
ARRAY_WIDTH = (PATCH_RADIUS * 2 + 1) * (PATCH_RADIUS * 2 + 1) * 40

# training datasets
trainA_df = pd.DataFrame(trainA)
trainA_CTC_df = trainA_df.loc[:,:ARRAY_WIDTH - 1]
trainA_AIF_df = trainA_df.loc[:,ARRAY_WIDTH:]
trainA_AIF_df.columns = range(40)
pca = PCA(n_components=5)
trainA_CTC_pca_df = pca.fit_transform(trainA_CTC_df)

# normalization attempts
train_group = trainA_CTC_df.groupby(lambda x: x % 40, axis=1)
trainA_sum = train_group.sum()
trainA_std = train_group.std()
trainA_median = train_group.median()

# validation dataset
verA_df = pd.DataFrame(verA)
verA_CTC_df = verA_df.loc[:,:ARRAY_WIDTH - 1]
verA_AIF_df = trainA_df.loc[:,ARRAY_WIDTH:]
verA_AIF_df.columns = range(40)
verA_CTC_pca_df = pca.fit_transform(verA_CTC_df)

# normalization attempts
ver_group = verA_CTC_df.groupby(lambda x: x % 40, axis=1)
verA_sum = ver_group.sum()
verA_std = ver_group.std()
verA_median = ver_group.median()

# PCA applied datasets
pca_t_df = pd.DataFrame(trainA_CTC_pca_df)
pca_v_df = pd.DataFrame(verA_CTC_pca_df)

# regularization
train_reg = trainA_df.copy(True)
for column in trainA_CTC_df.columns.values:
    train_reg.ix[:,column:column] = train_reg.ix[:,column:column] * trainA_AIF_df.ix[:,(column%40):(column%40)]

ver_reg = verA_df.copy(True)
for column in verA_CTC_df.columns.values:
    ver_reg.ix[:,column:column] = ver_reg.ix[:,column:column] * verA_CTC_df.ix[:,(column%40):(column%40)]
    
train_reg.fillna(train_reg.sum(), inplace=True)
ver_reg.fillna(ver_reg.sum(), inplace=True)
```

```python
larsModel = linear_model.LassoLars(alpha=.04, verbose=True)
larsModel.fit(trainA_CTC_df, trainb)
(larsModel.score(trainA_CTC_df, trainb), larsModel.score(verA_CTC_df, verb))
# (train, validation)
```

```python
larsModel = linear_model.LassoLars(alpha=.05, verbose=True)
larsModel.fit(trainA_sum, trainb)
(larsModel.score(trainA_sum, trainb), larsModel.score(verA_sum, verb))
# (train, validation)
```

```python
larsModel = linear_model.LassoLars(alpha=.06, verbose=True)
larsModel.fit(pca_t_df, trainb)
(larsModel.score(pca_t_df, trainb), larsModel.score(pca_v_df, verb))
# (train, validation)
```

```python
larsModel = linear_model.LassoLars(alpha=.045, verbose=True)
larsModel.fit(trainA_std, trainb)
(larsModel.score(trainA_std, trainb), larsModel.score(verA_std, verb))
# (train, validation)
```

```python
larsModel = linear_model.LassoLars(alpha=.10, verbose=True)
larsModel.fit(trainA_median, trainb)
(larsModel.score(trainA_median, trainb), larsModel.score(verA_median, verb))
# (train, validation)
```

```python
larsModel = linear_model.LassoLars(alpha=.05, verbose=True)
larsModel.fit(train_reg, trainb)
(larsModel.score(train_reg, trainb), larsModel.score(ver_reg, verb))
# (train, validation)
```
