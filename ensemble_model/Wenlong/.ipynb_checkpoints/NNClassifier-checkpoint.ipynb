```python
import time
import os

import pandas as pd
import numpy as np
import theano
import theano.tensor as T
import lasagne
```

```python
def load_dataset(filename):
    print(filename)
    df = pd.read_csv(filename)
    X_id = df.iloc[:, 0:4]
    X_train = df.iloc[:, 4:-1]
    y_train = df.iloc[:, -1:]
    return X_id, X_train, y_train
    #return np.matrix(X_id), np.matrix(X_train), np.matrix(y_train)
```

```python
X_id, X_train, y_train = load_dataset('./data/train_data.csv')
```

```python
# Helper Functions for NN
def iterate_minibatches(inputs, targets, batchsize, shuffle=False):
    assert len(inputs) == len(targets)
    if shuffle:
        indices = np.arange(len(inputs))
        np.random.shuffle(indices)
    for start_idx in range(0, len(inputs) - batchsize + 1, batchsize):
        if shuffle:
            excerpt = indices[start_idx:start_idx + batchsize]
            yield inputs[excerpt], targets[excerpt]
        else:
            excerpt = slice(start_idx, start_idx + batchsize)
            yield inputs[excerpt], targets[excerpt]
```

```python
def build_network(input_var = None):
    input_layer = lasagne.layers.InputLayer(shape=(None, 618),
                                            input_var=input_var)
    #input_layer = lasagne.layers.dropout(input_layer, p=0.2)
    #input_layer = lasagne.layers.batch_norm(input_layer)
    
    layer1 = lasagne.layers.DenseLayer(input_layer,
                                      num_units = 100,
                                      nonlinearity = lasagne.nonlinearities.leaky_rectify,
                                      W=lasagne.init.GlorotUniform())
    #layer = lasagne.layers.dropout(layer, p=0.5)
    #layer1 = lasagne.layers.batch_norm(layer1)
    
    
    layer2 = lasagne.layers.DenseLayer(layer1,
                                      num_units = 100,
                                      nonlinearity = lasagne.nonlinearities.leaky_rectify,
                                      W=lasagne.init.GlorotUniform())
    #layer = lasagne.layers.dropout(layer, p=0.5)
    #layer2 = lasagne.layers.batch_norm(layer2)
    
    '''
    layer = lasagne.layers.DenseLayer(layer,
                                      num_units = 50,
                                      nonlinearity = lasagne.nonlinearities.leaky_rectify,
                                      W=lasagne.init.GlorotUniform())
    layer = lasagne.layers.dropout(layer, p=0.5)
    
    layer = lasagne.layers.DenseLayer(layer,
                                      num_units = 100,
                                      nonlinearity = lasagne.nonlinearities.leaky_rectify,
                                      W=lasagne.init.GlorotUniform())
    layer = lasagne.layers.dropout(layer, p=0.5)
    '''
    
    output_layer = lasagne.layers.DenseLayer(layer2,
                                      num_units = 1,
                                      nonlinearity = lasagne.nonlinearities.sigmoid,
                                      W=lasagne.init.GlorotUniform())
    return output_layer

# Build neural network
input_var = T.matrix('inputs')
target_var = T.matrix('targets')

network = build_network(input_var)

output = lasagne.layers.get_output(network)
loss = lasagne.objectives.categorical_crossentropy(output, target_var).mean()
params = lasagne.layers.get_all_params(network, trainable=True)
#updates = lasagne.updates.nesterov_momentum(loss, params, learning_rate=0.001, momentum=0.9)
updates = lasagne.updates.adam(loss, params, learning_rate = 0.001, beta1=0.9, beta2=0.999, epsilon=1e-08)

prediction = lasagne.layers.get_output(network, deterministic=True)
accuracy = T.mean(T.eq(prediction, target_var), dtype=theano.config.floatX)

train_fn = theano.function([input_var, target_var], loss, updates=updates)
accuracy_fn = theano.function([input_var, target_var], accuracy)
```

```python
def get_data(X_train, y_train):
    X, X_val, y, y_val = train_test_split(X_train, np.ndarray.flatten(np.array(y_train)), test_size = 0.33)
    return X, y, X_val, y_val
```

```python
X_t, y_t, X_val, y_val = get_data(X_train, y_train)
y_t = np.reshape(np.matrix(y_t), (y_t.shape[0], 1))
y_val = np.reshape(np.matrix(y_val), (y_val.shape[0], 1))
#print(y_t.shape, y_val.shape)

num_epochs = 400

# Training Loop
print("Starting training...")
for epoch in range(num_epochs):
    # In each epoch, we do a full pass over the training data:
    train_err = 0
    train_batches = 0
    start_time = time.time()
    for batch in iterate_minibatches(X_t, y_t, 20, shuffle=True):
        inputs, targets = batch
        train_err += train_fn(inputs, targets)
        train_batches += 1

    # And a full pass over the validation data:
    val_acc = 0
    val_batches = 0
    for batch in iterate_minibatches(X_val, y_val, 20, shuffle=False):
        inputs, targets = batch
        acc = accuracy_fn(inputs, targets)
        val_acc += acc
        val_batches += 1

    # Then we print the results for this epoch:
    print("Epoch {} of {} took {:.3f}s".format(
        epoch + 1, num_epochs, time.time() - start_time))
    print("  training loss:\t\t{:.6f}".format(train_err / train_batches))
    print("  validation accuracy:\t\t{:.2f} %".format(
        val_acc / val_batches * 100))
```
